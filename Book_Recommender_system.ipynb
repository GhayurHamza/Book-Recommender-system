## Recommendation System

```python
#So we have 706 books and 810 users who have rated in the dataframe to be used for generating recommendations
# In the pt dataframe every book is a vector with 810 values
# we have to find similar books
# This can be done by finding the euclidean distance between books

# Importing cosine_similarity function to calculate distances
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Euclidean distance of 706 books with all the books in the form of an array
cosine_similarity(pt)
```

```python
# indicates values that provides similarity score of each book with respect to every book
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
#Function for generating recommendations
#input is book_name and output will be recommendations
def recommend(book_name):
    # fetching the index of the book entered
    index = np.where(pt.index==book_name)[0][0]
    
    #array of similarity scores of the book searched for by using
    #its index
    #it is enumerated to include the index with data
    #and revered to get the books which have highest similarity score
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    
    return data
```

```python
recommend('Fahrenheit 451')
```
